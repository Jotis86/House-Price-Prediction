```python
# Machine Learning Model
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, mean_absolute_percentage_error, explained_variance_score
```

```python
# Select features and target variable
X = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms', 'Avg. Area Number of Bedrooms', 'Area Population']]
y = df['Price']
```

```python
# Normalize and scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Split the scaled data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```python
# Create the linear regression model
model = LinearRegression()
```

```python
# Train the model
model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = model.predict(X_test)
```

```python
# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
mape = mean_absolute_percentage_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
explained_variance = explained_variance_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'Root Mean Squared Error: {rmse}')
print(f'Mean Absolute Error: {mae}')
print(f'Mean Absolute Percentage Error: {mape}')
print(f'R^2 Score: {r2}')
print(f'Explained Variance Score: {explained_variance}')
```

```python
# Visualize the predictions vs actual values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linewidth=2)
plt.title('Predictions vs Actual Values')
plt.xlabel('Actual Values')
plt.ylabel('Predictions')
plt.show()
```

```python
# Residual plot
plt.figure(figsize=(10, 6))
sns.residplot(x=y_test, y=y_pred, lowess=True, color='purple')
plt.title('Residuals Plot')
plt.xlabel('Actual Values')
plt.ylabel('Residuals')
plt.show()
```

```python
# Distribution of residuals
plt.figure(figsize=(10, 6))
sns.histplot(y_test - y_pred, kde=True, color='green')
plt.title('Distribution of Residuals')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot of errors
errors = y_test - y_pred
plt.figure(figsize=(10, 6))
plt.scatter(y_test, errors, alpha=0.5)
plt.axhline(y=0, color='red', linestyle='--')
plt.title('Errors vs Actual Values')
plt.xlabel('Actual Values')
plt.ylabel('Errors')
plt.show()
```

```python
# Plot of absolute percentage errors
percentage_errors = np.abs(errors / y_test) * 100
plt.figure(figsize=(10, 6))
plt.scatter(y_test, percentage_errors, alpha=0.5)
plt.axhline(y=np.mean(percentage_errors), color='red', linestyle='--')
plt.title('Percentage Errors vs Actual Values')
plt.xlabel('Actual Values')
plt.ylabel('Percentage Errors')
plt.show()
```

```python
def predict_house_price(avg_area_income, avg_area_house_age, avg_area_number_of_rooms, avg_area_number_of_bedrooms, area_population):
    # Create a DataFrame for the new input values
    new_data = pd.DataFrame({
        'Avg. Area Income': [avg_area_income],
        'Avg. Area House Age': [avg_area_house_age],
        'Avg. Area Number of Rooms': [avg_area_number_of_rooms],
        'Avg. Area Number of Bedrooms': [avg_area_number_of_bedrooms],
        'Area Population': [area_population]
    })
    
    # Normalize and scale the new input values
    new_data_scaled = scaler.transform(new_data)
    
    # Predict the house price
    predicted_price = model.predict(new_data_scaled)
    
    return predicted_price[0]
```

```python
# Example: Predict house price for new input values
avg_area_income = 65000
avg_area_house_age = 5
avg_area_number_of_rooms = 7
avg_area_number_of_bedrooms = 4
area_population = 30000

predicted_price = predict_house_price(avg_area_income, avg_area_house_age, avg_area_number_of_rooms, avg_area_number_of_bedrooms, area_population)
print(f'Predicted House Price: ${predicted_price:.2f}')
```
